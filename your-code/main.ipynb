```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import numpy as np

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
#First we create a table with the data we need with groupby

barchart = data.groupby("ItemType")[["RetailSales", "RetailTransfers", "WarehouseSales"]].sum()
```

```python
barchart
```

```python
ax = barchart.plot(kind = "bar", figsize = (12,10))
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
data["TotalSales"] = data["RetailSales"] + data["RetailTransfers"] + data["WarehouseSales"]
chartdata = data.groupby("Supplier")[["TotalSales"]].sum().sort_values(by="TotalSales", ascending=False).head(10)
```

```python
chartdata
```

```python
chartdata.plot(kind="barh")
plt.show()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
linechart = data.groupby(["Year","Month"])[["RetailSales", "RetailTransfers", "WarehouseSales"]].mean()
```

```python
linechart
```

```python
linechart.plot(kind="line", figsize = (12,10))
plt.show()
```

## 4. Plot the same information as above but as a bar chart.

```python
linechart.plot(kind="bar")
plt.show()
```
